```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('/content/drive/MyDrive/mi_twitter_data_raw.csv')

# Mantener solo las columnas 'Full_Text' y 'MI_Type'
df_filtered = df[['Full_Text', 'MI_Type']]

# Renombrar las columnas
df_filtered.rename(columns={'Full_Text': 'statement', 'MI_Type': 'status'}, inplace=True)

# Convertir los valores de la columna 'status' a capitalización inicial
df_filtered['status'] = df_filtered['status'].str.capitalize()

# Contar el número de registros en cada categoría de la columna 'status'
category_counts = df_filtered['status'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Número de registros en cada categoría de status')
plt.xlabel('Categoría')
plt.ylabel('Número de registros')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar el gráfico
plt.show()

# Guardar el nuevo DataFrame en un archivo CSV en el entorno de Colab
output_path = '/content/mi_twitter_data_filtered.csv'
df_filtered.to_csv(output_path, index=False)

# Descargar el archivo a tu computadora local
#files.download(output_path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Fijar la semilla para obtener resultados reproducibles
np.random.seed(42)

# Cargar todos los archivos de datos
df_base = pd.read_csv('/content/drive/MyDrive/Combined Data.csv')
df_mi_twitter = pd.read_csv('/content/mi_twitter_data_filtered.csv')
df_twitter_full = pd.read_csv('/content/Twitter_Full_filtered.csv')
df_suicide_ideation = pd.read_csv('/content/Suicide_Ideation_Dataset_filtered.csv')
df_twitter_suicidal = pd.read_csv('/content/twitter_suicidal_data_filtered.csv')

# Concatenar todos los DataFrames en uno solo
df_combined = pd.concat([df_base, df_mi_twitter, df_twitter_full, df_suicide_ideation, df_twitter_suicidal])

# Definir el número máximo de registros por categoría
max_records = 16000

# Función para equilibrar las categorías
def balance_category(df, category, max_records):
    category_df = df[df['status'] == category]
    if category_df.shape[0] > max_records:
        return category_df.sample(max_records, replace=False)
    else:
        return category_df

# Aplicar la función a cada categoría y concatenar los resultados
categories = df_combined['status'].unique()
balanced_df = pd.concat([balance_category(df_combined, category, max_records) for category in categories])

# Eliminar registros duplicados basados en el contenido de la columna 'statement'
balanced_df = balanced_df.drop_duplicates(subset=['statement'])

# Mezclar aleatoriamente los registros y reiniciar el índice
balanced_df = balanced_df.sample(frac=1).reset_index(drop=True)

# Guardar el DataFrame final en un archivo CSV
output_path = '/content/Combined_Balanced_Data.csv'
balanced_df.to_csv(output_path, index=False)

# Descargar el archivo a tu computadora local
files.download(output_path)

# Crear una gráfica del número de registros por categoría
category_counts = balanced_df['status'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Número de registros en cada categoría después del balanceo')
plt.xlabel('Categoría')
plt.ylabel('Número de registros')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
